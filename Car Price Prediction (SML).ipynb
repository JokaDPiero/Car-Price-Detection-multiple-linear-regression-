```python
df1=df.copy()
```

```python
#Encoding for multiple linear Regression
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
df1['fueltype']=enc.fit_transform(df1['fueltype'])
df1['aspiration']=enc.fit_transform(df1['aspiration'])
df1['doornumber']=enc.fit_transform(df1['doornumber'])
df1['CarName']=enc.fit_transform(df1['CarName'])
df1['carbody']=enc.fit_transform(df1['carbody'])
df1['drivewheel']=enc.fit_transform(df1['drivewheel'])
df1['cylindernumber']=enc.fit_transform(df1['cylindernumber']) 
df1['enginelocation']=enc.fit_transform(df1['enginelocation'])
df1['enginetype']=enc.fit_transform(df1['enginetype'])
df1['fuelsystem']=enc.fit_transform(df1['fuelsystem'])
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df1.corr(),annot=True,cmap="RdYlGn")
```

Here,price and engine size have correation of 0.87 and are highly correlated.

```python
x=df1.drop('price',axis="columns")
y=df1['price']
```

```python
#Creating training and testing sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=10,test_size=0.2)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#Training our sets in the model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred=pd.DataFrame(y_pred,columns=['Predicted Price'])
y_pred
```
